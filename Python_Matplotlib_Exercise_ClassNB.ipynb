### Load Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Load company sales data and store to `sales`

```python
sales=pd.read_csv('company_sales_data.csv')
sales.head()
```

## Use pandas to impute missing valies if there is any

```python
sales.isnull().sum()
```

```python
shirt=sales['Shirt'].mode()[0]
sales['Shirt'].fillna(shirt, inplace=True)
```

```python
print(sales.isnull().sum())
print('')
print('I filled the null value with the mode of that column')
```

## Exercise 1.Read Bags, Top, Shirt and Pant sales data and show it  using a multiline plot


```python
fig=plt.figure(figsize=(12,5))

plt.plot(sales['month_number'],sales['Bags'],
        label='Bag Sales Data',
        marker='o')

plt.plot(sales['month_number'],sales['Top'],
        label='Top Sales Data',
        marker='o')

plt.plot(sales['month_number'],sales['Shirt'],
        label='Shirt Sales Data',
        marker='o')

plt.plot(sales['month_number'],sales['Pant'],
        label='Pant Sales Data',
        marker='o')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.show()
```

## Exercise 2: Read Pant and Shirt sales data of each month and show it using a scatter plot


```python
plt.scatter(sales['month_number'],sales['Pant'],
           color='red',
           label='Pant sales data')

plt.scatter(sales['month_number'],sales['Shirt'],
           color='blue',
           label='Pant sales data')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.show()
```

## Exercise 3: Read Jeans and Top product sales data and show it using the bar chart

Note:
-  move x axis of jeans .25 left and Top .25 right and add width
-  plt.bar(sales['month_number']-.25, sales['Jeans'],color='red',width=.30, label="Jeans sales data")

```python
plt.bar(sales['month_number']-.25, sales['Jeans'],
       color='red',
       label='Jeans sales data',
       width=.30)

plt.bar(sales['month_number']+.25, sales['Top'],
       color='blue',
       label='Top sales data',
       width=.30)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.show()
```

## Exercise 4: Read the total profit of each month and show it using the histogram to see the most common profit ranges 
The histogram should look like this:


```python
plt.hist(sales['total_profit'],
        label='Profit data')

plt.xlabel("Profit in dollar")
plt.legend(loc='upper right')
plt.show()
```

## Exercise 5: Calculate `total sale` data for last year for each product and show it using a Pie chart

- salesData=[sales['Bags'].sum(),.......]

```python
salesData=[sales['Bags'].sum(),
           sales['Tie'].sum(),
           sales['Shirt'].sum(),
           sales['Jeans'].sum(),
           sales['Top'].sum(),
           sales['Pant'].sum()]
column_names=['Bags','Tie','Shirt','Jeans','Top','Pant']
```

```python
plt.pie(salesData,
       labels=column_names,
       autopct='%.1f%%')

plt.axis('equal')
plt.show()
```

### Exercise 6: Read all product sales data and show it using the stack plot. Save the result as `stack_sales.png`


```python
sales.columns
```

```python
fig=plt.stackplot(sales['month_number'],
              sales['Bags'],
             sales['Tie'],
             sales['Shirt'],
             sales['Jeans'],
             sales['Top'],
             sales['Pant'],
             labels=column_names)

plt.xlabel('Month Number')
plt.legend(loc='upper left')
plt.show()
```
